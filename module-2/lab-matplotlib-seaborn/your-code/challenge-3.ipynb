#### Use the appropriate plot to visualize the column `Pclass` 


```python
# your code here
def countclass():
    count=[0,0,0]
    for e in titanic['Pclass']:
        if e==1:
            count[0]+=1
        elif e==2:
            count[1]+=1
        else:
            count[2]+=1
    return count
        

plt.bar([1,2,3],countclass())
```

#### Use the appropriate plot to display the summary statistics of the `Age` in function of the `Pclass`.
